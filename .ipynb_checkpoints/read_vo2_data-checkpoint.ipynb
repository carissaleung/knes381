```python
import pandas as pd
import matplotlib.pyplot as plt

skiprow = list(range(0,))
df = pd.read_csv('VO2_CSV.csv', header=[0], skiprows=[1,2, 3])


# df = pd.read_excel('VO2_excel.xlsx', header=[0,2], skiprows=[3]) #this line works
# header=[0, 2] reads the header from rows 1 and 3, header=[0,1,2] reads header info in row 1 2 and 3
# skiprows=2 skips first 2 rows skiprows=[2,4] skips the specific rows 2 and 4, skiprows=3 skips first three rows
#df.drop(2, axis=0)
df.head()


#df.plot(x="TIME", y="VO2")
#plt.show()
```

```python
df.info()
```

```python
x = df['TIME ']
y = df['VO2   ']
# y1 = df['VE     ']

# https://matplotlib.org/stable/index.html
ymax = max(y)
xmax = x[y.argmax()]
ax = plt.subplot()
ax.annotate('$\dot VO_2max$ =({}) L/min'.format(round(ymax, 2)), xy=(xmax, ymax), xytext=(xmax, ymax+.25),
            arrowprops=dict(facecolor='red'),
            )


ax.plot(x, y,  label=('$\dot VO_2$ L/min') )
# ax.plot(x, y1,  label=('$\dot VCO_2$ L/min') )
ax.spines[['right', 'top']].set_visible(False)
ax.set(xlabel=('time'), title = ('$\dot VO_2$'), label=('$\dot VO_2$ L/min'))
ax.set(ylabel=('L/min'))
ax.legend()

# Code below works without the fig, ax settings...
#df.plot(x =0, y =1, xlabel='Time (min)', ylabel=(r'$\dot VO_2 (L/min)$'), label=('$\dot VO_2$ L/min'), 
#     title='Simple $\dot VO_2$ time plot')
```
